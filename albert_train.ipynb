### Preprocessing

Albert requires input data be formatted into `train.tsv`, `dev.tsv`, and `test.tsv` files.  This has been done in `convert_dataset.ipynb`.  The files have been saved in the `data/` foler.

The Albert base pre-trained model comes with a vocabulary file, but one was made from Tweets.  See `make_vocab.ipynb` for those steps.

### TX2 Compatibility

`ALBERT/run_classifier.py` has been modified for TX2 compatibility.  More information about this is can be found in `results.ipynb`.

### Resource Restrictions and Notebook Limitations

The commands below were ran in Terminal, not this notebook, because the massive quantity of text output will crash Jupyter.  If you want to run these commands within this notebook, you'll need to disable logging.

Because of limited ram on the TX2 is is recommended all kernals are shut down or restarted before running any of the commands below.

You'll want at least 4 GB of ram free before running.  This can be checked by running in `jtop` in terminal.
